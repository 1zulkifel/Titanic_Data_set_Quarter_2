```python
import numpy as np
import pandas as pd
```

```python
df1=pd.DataFrame(np.random.randint(1,33, size=(1000,5)),columns=['s1','s2','s3','s4','s5',])
df2=pd.DataFrame(np.random.randint(34,40, size=(1000,5)),columns=['s1','s2','s3','s4','s5',])
df3=pd.DataFrame(np.random.randint(41,50, size=(1000,5)),columns=['s1','s2','s3','s4','s5',])
df4=pd.DataFrame(np.random.randint(51,60, size=(1000,5)),columns=['s1','s2','s3','s4','s5',])
df5=pd.DataFrame(np.random.randint(61,70, size=(1000,5)),columns=['s1','s2','s3','s4','s5',])
df6=pd.DataFrame(np.random.randint(71,80, size=(1000,5)),columns=['s1','s2','s3','s4','s5',])
df7=pd.DataFrame(np.random.randint(81,100, size=(1000,5)),columns=['s1','s2','s3','s4','s5',])

df=pd.concat([df1,df2,df3,df4,df5,df6,df7])
df.head()
```

```python
df['Obtained']=df.loc[:,'s1':'s5'].apply(sum,axis=1)
df.head()
```

```python
df['Total']=500
df
```

```python
df['Percentage']=df.Obtained/df.Total*100
df.tail()
```

```python
def grade(x):
    if x>=80:
        return 'A+'
    elif x>=70:
        return 'A'
    elif x>=60:
        return 'B'
    elif x>=50:
        return 'C'
    elif x>=40:
        return 'D'
    elif x>=33:
        return 'E'
    else:
        return 'Fail'
df['Grade']=df.Percentage.apply(grade)
df.head()
```

```python
df.head()
```

```python
df.Grade.value_counts()
```

## Apply vectorization on Y

```python
#see the relation
Y=df[['Grade']]
Y=pd.get_dummies(Y)
Y.head()
```

```python
Y=df['Grade']
Y=pd.get_dummies(Y)
Y.head()
```

```python
Y.drop_duplicates()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam,SGD
```

```python
model=Sequential()
model.add(Dense(9,activation='relu', input_shape=(5,)))
model.add(Dense(7,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer= SGD(lr=0.1),
              loss='categorical_crossentropy',
              metrics=['acc'])
```

```python
model.fit(X,Y,epochs=5)
```

```python
df['predict_Y']=Y.columns.take(np.argmax(model.predict(X),
                              axis=1))
Y.head()
```
